```python
import numpy as np
import matplotlib.pyplot as plt

# Data points
x = np.array((10, 100, 1000, 10000, 100000, 1000000, 10000000, 100000000))
y = np.array((0, 0, 0, 3014000, 6047000, 60420000, 472104000, 4425808000))

# Logarithmic function
log10_y = np.log10(x)  # Base-10 log

# Plotting both graphs
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Original Data", marker="o", linestyle="-", color="blue")
plt.plot(x, log10_y, label="log10(x) (Base-10 Log)", marker="^", linestyle=":", color="green")

# Setting log scale for better visualization (Optional)
plt.xscale("log")  # Log scale for x-axis
plt.yscale("log")  # Log scale for y-axis (optional, based on data)

# Labels and title
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Comparison of Data with Logarithmic Functions")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data points
x = np.array((10, 100, 1000, 10000, 100000, 1000000, 10000000, 100000000))
y = np.array((0, 0, 0, 3014000, 6047000, 60420000, 472104000, 4425808000))

# Logarithmic function
# log_y = np.log(x)  # Natural log (ln)
log10_y = np.log10(x)  # Base-10 log

# Plotting both graphs
plt.figure(figsize=(6, 6))
plt.plot(x, y, label="Original Data", marker="o", linestyle="-", color="blue")
plt.plot(x, x, label="x", marker="s", linestyle="--", color="red")
plt.plot(x, log10_y, label="log10(x) (Base-10 Log)", marker="^", linestyle=":", color="green")

# Setting log scale for better visualization (Optional)
plt.xscale("log")  # Log scale for x-axis
plt.yscale("log")  # Log scale for y-axis (optional, based on data)

# Labels and title
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Comparison of Data with Logarithmic Functions")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

# Show plot
plt.show()
```
