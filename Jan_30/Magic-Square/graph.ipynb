```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("magic_square.csv")

size = df["size"]
time1 = df["Time(ms)"]
# time1 - Magic Square time
# time2 - n^2

time2 = np.array(size) ** 2
time2 = time2 * (time1.max() / time2.max())

plt.figure(figsize=(10, 5))

# Plotting Magic Square
plt.plot(size, time1, label="Magic Square Time", color="b")

# Plotting O(n) Curve
plt.plot(size, time2, label="O(n)", color="r", linestyle="dashed")

plt.xlabel("Size")
plt.ylabel("Time (milliseconds)")
plt.title("Magic Square Time vs O(n) Complexity")
plt.legend()
plt.grid(True)

plt.show()
```
