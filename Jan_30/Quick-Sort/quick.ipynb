```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("Quickdata.csv")

size = df["Size"]
time1 = df["Time (nanoseconds)"]
# time1 - quicksort time
# time2 - n^2

time2 = np.array(size) ** 2
time2 = time2 * (time1.max() / time2.max())

plt.figure(figsize=(10, 5))

# Plotting QuickSort
plt.plot(size, time1, label="QuickSort Time", color="b")

# Plotting O(n²) Curve
plt.plot(size, time2, label="O(n²)", color="r", linestyle="dashed")

plt.xlabel("Size")
plt.ylabel("Time (nanoseconds)")
plt.title("QuickSort Time vs O(n²) Complexity")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
df = pd.read_csv("Quickdata.csv")

# Extract size and time data
size = df["Size"]
time1 = df["Time (nanoseconds)"]  # QuickSort execution time

# Compute O(n log n) curve
time2 = np.array(size) * np.log(np.array(size))
time2 = time2 * (time1.max() / time2.max())  # Normalize to match scale

# Plotting
plt.figure(figsize=(10, 5))

# QuickSort Execution Time
plt.plot(size, time1, label="QuickSort Time", color="b")

# O(n log n) Complexity Curve
plt.plot(size, time2, label="O(n log n)", color="r", linestyle="dashed")

# Labels and Title
plt.xlabel("Input Size (n)")
plt.ylabel("Time (nanoseconds)")
plt.title("QuickSort Time vs O(n log n) Complexity")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
