```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("MergeData.csv")

size = df["Size"]
time1 = df["Time (ns)"]
# time1 - mergesort time
# time2 - n^2

time2 = np.array(size) ** 2
time2 = time2 * (time1.max() / time2.max())

plt.figure(figsize=(10, 5))

# Plotting Mergesort
plt.plot(size, time1, label="Mergesort Time", color="b")

# Plotting O(n) Curve
plt.plot(size, time2, label="O(n)", color="r", linestyle="dashed")

plt.xlabel("Size")
plt.ylabel("Time (nanoseconds)")
plt.title("Mergesort Time vs O(n) Complexity")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
df = pd.read_csv("bestcase.csv")

# Extract size and actual Mergesort time
size = df["Size"].to_numpy()
time1 = df["Time (ns)"].to_numpy()  # Mergesort actual timings

# Calculate O(n log n) theoretical values
time2 = size * np.log2(size)

# Normalize theoretical to actual scale for comparison
time2 = time2 * (time1.max() / time2.max())

# Calculate relative error
relative_error = np.abs(time1 - time2) / time2

# Filter: keep only points where relative error < 20%
threshold = 0.20
mask = relative_error < threshold

filtered_size = size[mask]
filtered_time1 = time1[mask]
filtered_time2 = time2[mask]

# Plotting
plt.figure(figsize=(10, 5))

plt.plot(filtered_size, filtered_time1, label=" Mergesort Time", color="b", linewidth=2)
plt.plot(filtered_size, filtered_time2, label="Ideal O(n log n)", color="r", linestyle="dashed", linewidth=2)

plt.xlabel("Size")
plt.ylabel("Time (nanoseconds)")
plt.title(" Mergesort Time vs O(n log n) Complexity")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
