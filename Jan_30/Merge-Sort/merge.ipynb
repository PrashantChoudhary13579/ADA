```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("MergeData.csv")

size = df["Size"]
time1 = df["Time (ns)"]
# time1 - mergesort time
# time2 - n^2

time2 = np.array(size) ** 2
time2 = time2 * (time1.max() / time2.max())

plt.figure(figsize=(10, 5))

# Plotting Mergesort
plt.plot(size, time1, label="Mergesort Time", color="b")

# Plotting O(n²) Curve
plt.plot(size, time2, label="O(n²)", color="r", linestyle="dashed")

plt.xlabel("Size")
plt.ylabel("Time (nanoseconds)")
plt.title("Mergesort Time vs O(n²) Complexity")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("MergeData.csv")

size = df["Size"]
time1 = df["Time (ns)"]
# time1 - mergesort time
# time2 - n^2
# time3 - n log n
# time4 - n² log n

# O(n²) time complexity
time2 = np.array(size) ** 2
time2 = time2 * (time1.max() / time2.max())

# O(n log n) time complexity
time3 = size * np.log2(size)
time3 = time3 * (time1.max() / time3.max())

# O(n² log n) time complexity
time4 = (size ** 2) * np.log2(size)
time4 = time4 * (time1.max() / time4.max())

plt.figure(figsize=(10, 5))

# Plotting Mergesort
plt.plot(size, time1, label="Mergesort Time", color="b")

# Plotting O(n²) Curve
plt.plot(size, time2, label="O(n²)", color="r", linestyle="dashed")

# Plotting O(n log n) Curve
plt.plot(size, time3, label="O(n log n)", color="g", linestyle="dotted")

# Plotting O(n² log n) Curve
plt.plot(size, time4, label="O(n² log n)", color="orange", linestyle="dashdot")

plt.xlabel("Size")
plt.ylabel("Time (nanoseconds)")
plt.title("Mergesort Time vs Time Complexity Curves")
plt.legend()
plt.grid(True)

plt.show()

```
