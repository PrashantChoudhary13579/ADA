```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("bubbleData.csv")

size = df["size"]
time1 = df["Time(ns)"]
# time1 - Bubble sort time
# time2 - n^2

time2 = np.array(size) ** 2
time2 = time2 * (time1.max() / time2.max())

plt.figure(figsize=(10, 5))

# Plotting Bubble sort
plt.plot(size, time1, label="Bubble sort Time", color="b")

# Plotting O(n²) Curve
plt.plot(size, time2, label="O(n²)", color="r", linestyle="dashed")

plt.xlabel("Size")
plt.ylabel("Time (nanoseconds)")
plt.title("Bubble sort Time vs O(n²) Complexity")
plt.legend()
plt.grid(True)

plt.show()
```
