```python
# worst and average case
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("bubbleData.csv")

size = df["size"]
time1 = df["Time(ns)"]
# time1 - Bubble sort time
# time2 - n^2

time2 = np.array(size) ** 2
time2 = time2 * (time1.max() / time2.max())

plt.figure(figsize=(10, 5))

# Plotting Bubble sort
plt.plot(size, time1, label="Bubble sort Time", color="b")

# Plotting O(n) Curve
plt.plot(size, time2, label="O(n)", color="r", linestyle="dashed")

plt.xlabel("Size")
plt.ylabel("Time (nanoseconds)")
plt.title("Bubble sort Time vs O(n) Complexity")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# best case
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("bestcase.csv")

size = df["size"]
time1 = df["Time(ns)"]
# time1 - Bubble sort time
# time2 - n^2

time2 = np.array(size) ** 2
time2 = time2 * (time1.max() / time2.max())

plt.figure(figsize=(10, 5))

# Plotting Bubble sort
plt.plot(size, time1, label="Bubble sort Time", color="b")

# Plotting O(n) Curve
plt.plot(size, time2, label="O(n)", color="r", linestyle="dashed")

plt.xlabel("Size")
plt.ylabel("Time (nanoseconds)")
plt.title("Bubble sort Time vs O(n) Complexity")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Input sizes (same as before)
size = np.arange(10, 170*100 + 10, 100)

# Space used is constant (e.g., 4 units)
space_used = np.full_like(size, fill_value=4)

plt.figure(figsize=(10, 5))
plt.plot(size, space_used, label="Space Complexity O(1)", color="green")

plt.xlabel("Input Size (n)")
plt.ylabel("Space Used (units)")
plt.title("Space Complexity of Bubble Sort")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
