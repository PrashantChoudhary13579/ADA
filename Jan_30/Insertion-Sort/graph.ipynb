```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("insertionData.csv")

size = df["Size"]
time1 = df["Time(ns)"]
# time1 - insertion sort time
# time2 - n^2

time2 = np.array(size) ** 2
# time3 = np.log2(size)
time2 = time2 * (time1.max() / time2.max())
# time3 = time3 *(time1.max()/time3.max())

plt.figure(figsize=(10, 5))

# Plotting Insertion Sort
plt.plot(size, time1, label="Insertion Sort Time", color="b")

# Plotting O(n) Curve
plt.plot(size, time2, label="O(n)", color="r", linestyle="dashed")

# plt.plot(size,time3,label="min",color ="green")

plt.xlabel("Size")
plt.ylabel("Time (nanoseconds)")
plt.title("Insertion Sort Time vs O(n) Complexity")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
df = pd.read_csv("bestcase.csv")

size = df["Size"]
time1 = df["Time(us)"]

# Theoretical O(n) line
theoretical = np.array(size)
theoretical = theoretical * (time1.max() / theoretical.max())

# Set threshold: allow only points where time is close to O(n) line
threshold_ratio = 1.20  # Keep points within 20% above the O(n) line
mask = (time1 >= theoretical) & (time1 <= theoretical * threshold_ratio)

# Filter data
filtered_size = size[mask]
filtered_time = time1[mask]

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(filtered_size, filtered_time, label="Filtered Insertion Sort Times", color="blue", marker='o', linestyle='-')
plt.plot(size, theoretical, label="O(n)", color="red", linestyle="dashed")

plt.xlabel("Size")
plt.ylabel("Time (microseconds)")
plt.title("Filtered Insertion Sort (Best Case) Near O(n)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
df = pd.read_csv("bestcase.csv")

size = df["Size"]
time1 = df["Time(us)"]

# Create O(n) reference line
on_line = np.array(size)
on_line = on_line * (time1.max() / on_line.max())

# Calculate absolute difference from O(n)
diff = np.abs(time1 - on_line)

# Keep only points above the O(n) line
mask_above = time1 >= on_line
size_above = size[mask_above]
time_above = time1[mask_above]
diff_above = diff[mask_above]

# Get indices of the 10 points with smallest difference above the line
top_10_indices = diff_above.nsmallest(10).index

# Filter top 10 close-to-O(n) points
filtered_size = size.loc[top_10_indices]
filtered_time = time1.loc[top_10_indices]

# Plot
plt.figure(figsize=(10, 5))
plt.plot(filtered_size, filtered_time, label="Top 10 Points Near O(n)", color="blue", marker='o', linestyle='-')
plt.plot(size, on_line, label="O(n)", color="red", linestyle="dashed")

plt.xlabel("Size")
plt.ylabel("Time (microseconds)")
plt.title("Best Case Insertion Sort Time vs O(n) - Top 10 Close Points")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
