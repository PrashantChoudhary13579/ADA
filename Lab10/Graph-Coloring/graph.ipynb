```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read the CSV file
data = pd.read_csv('complexity.csv')

# Assuming m is fixed (e.g., 3 colors)
m = 3

# Calculate theoretical time (m^n) and normalize it for plotting
data['theoretical'] = [m**n for n in data['nodes']]
# Normalize to match actual runtime scale
scaling_factor = data['time_seconds'].iloc[-1] / data['theoretical'].iloc[-1]
data['theoretical_scaled'] = data['theoretical'] * scaling_factor

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(data['nodes'], data['time_seconds'], marker='o', label='Actual Time', color='blue')
plt.plot(data['nodes'], data['theoretical_scaled'], marker='s', linestyle='--', label=f'Theoretical O({m}^n)', color='red')

# Labels and title
plt.title('Graph Coloring: Actual vs Theoretical Time Complexity')
plt.xlabel('Number of Nodes (n)')
plt.ylabel('Time (seconds)')
plt.legend()
plt.grid(True)
plt.xticks(data['nodes'])  # Show exact n values
plt.tight_layout()

# Save and show plot
plt.savefig('graph_coloring_comparison.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read the CSV file
data = pd.read_csv('complexity.csv')

# Number of colors
m = 3
log_m = np.log10(m)

# Compute log(time) and log(theoretical)
data['log_time'] = np.log10(data['time_seconds'] + 1e-10)
data['log_theoretical'] = data['nodes'] * log_m

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(data['nodes'], data['log_time'], label='Log(Actual Time in seconds)', marker='o', color='blue')
plt.plot(data['nodes'], data['log_theoretical'], label='Log(Theoretical Time O(3^n))', linestyle='--', color='red')

plt.xlabel('Number of Nodes (n)')
plt.ylabel('Log(Time)')
plt.title('Graph Coloring: Actual vs Theoretical Time Complexity (Log Scale)')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save and show
plt.savefig('graph_coloring_log_comparison.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read data
data = pd.read_csv('complexity.csv')

# Number of colors
m = 3

# Compute theoretical time (e.g., 3^n) safely using log scale directly
data['theoretical'] = data['nodes'].apply(lambda n: m ** n)

# Normalize both actual and theoretical for better visual comparison
data['normalized_time'] = data['time_seconds'] / data['time_seconds'].max()
data['normalized_theoretical'] = data['theoretical'] / data['theoretical'].max()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(data['nodes'], data['normalized_time'], label='Normalized Actual Time', marker='o', color='blue')
plt.plot(data['nodes'], data['normalized_theoretical'], label='Normalized 3^n (Theoretical)', linestyle='--', color='red')

plt.xlabel('Number of Nodes (n)')
plt.ylabel('Normalized Values')
plt.title('Graph Coloring: Actual vs Theoretical Complexity (Normalized)')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save and show
plt.savefig('graph_coloring_normalized_comparison.png', dpi=300)
plt.show()
```
