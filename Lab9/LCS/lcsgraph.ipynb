```python
import matplotlib.pyplot as plt
import pandas as pd

# Read data from CSV
df = pd.read_csv("string.csv")

# Extract x and y
lengths = df['length']
times = df['time_in_seconds']

# Generate m*n (quadratic growth) reference line (scaled for comparison)
quadratic = [(n ** 2) / (10**7) for n in lengths]  # scale down for visibility

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(lengths, times, label='LCS Time (seconds)', marker='o', color='blue')
plt.plot(lengths, quadratic, label='Reference: n² (scaled)', linestyle='--', color='red')

plt.title("LCS Time Complexity vs. String Length")
plt.xlabel("Length of Strings")
plt.ylabel("Execution Time (seconds)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
